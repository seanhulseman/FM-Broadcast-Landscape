```python
us_states['area']=0

# convert to Albers Equal Area crs epsg=5070 
states_albers = us_states.to_crs(epsg=5070)
for i in states_albers['state']:
    sq_mi = states_albers.loc[states_albers['state']==i].geometry.area*3.861E-7 
    states_albers['area'].loc[states_albers['state']==i]=sq_mi
states_albers.head()
```

```python
#calculate total stations per area in each state
states_albers['station_area_ratio']=0
for i in states_albers['state']:
    ratio = states_albers['total_stations'].loc[states_albers['state']==i]/states_albers['area'].loc[states_albers['state']==i]
    states_albers['station_area_ratio'].loc[states_albers['state']==i]=ratio
states_albers.head()
```

```python
#calculate total stations per person in state
states_albers['station_populatio_ratio']=0
for i in states_albers['state']:
    ratio = states_albers['total_stations'].loc[states_albers['state']==i]/states_albers['population'].loc[states_albers['state']==i]
    states_albers['station_populatio_ratio'].loc[states_albers['state']==i]=ratio
states_albers.head()
```
